## Calculate OPS

* Insert table column to the right of SH named PA for Plate Appearances
    * $PA = AB + BB + HBP + SF$
    
![alt text](images/create_pa.JPG)

* Insert table column to the right of PA named OBP for On-Base Percentage
    * $OBP = (H + BB + HBP) / PA$
    
![alt text](images/create_obp.JPG)

* Insert table column to the right of OBP named SLG for Slugging Percentage
Create 

    * $SLG = ((H-(2B+3B+HR)) + (2 * 2B) + (3 * 3B) + (4 * HR)) / AB$
    
![alt text](images/create_slg.JPG)

* Insert table column to the right of SLG named OPS for On-Base plus Slugging
    * $OPS = OBP + SLG$
    
 ![alt text](images/create_ops.JPG)


* Sort the values from largest to smallest to get the top 5 seasons by OPS of these players from 2012-2015

 ![alt text](images/top_ops.JPG)
